##### UNIVARIATE ANALYSYS

```python
def num_plot_dist(df,columns):
    for col in columns:
        plt.figure(figsize=(15,7))

        plt.subplot(1, 2, 1)
        sns.histplot(df[col],kde=True,palette='Blues')
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('frequency')

        plt.subplot(1, 2, 2)
        plt.boxplot(x=df[col],vert=False)
        plt.title(f'boxplot of {col}')
        plt.xlabel(col)

        plt.tight_layout()
        plt.show()

```

```python
num_plot_dist(data,num_cols)
```

```python
def cat_plot(df,columns):
    for col in columns:
        plt.figure(figsize=(10,6))
        sns.countplot(x=col,data=df,order=df[col].value_counts().index)
        plt.title(f"Count plot for {col}")
        plt.xlabel(col)
        plt.ylabel("frequency")
        plt.show()
```

```python
cat_plot(data,cat_cols)
```

#### BIVARIATE GRAPHS

```python
def plot_bivariate_num(target,columns,df):
    for col in columns:
        plt.figure(figsize=(10,6))
        sns.boxplot(y=col,x=target,data=df,palette='Blues')
        plt.xlabel(target)
        plt.ylabel(col)
        plt.title(f"{col} vs {target}")
        plt.tight_layout()
        plt.show()
```

```python
plot_bivariate_num('booking_status',num_cols,data)
```

```python
def plot_bivariate_cat(data,columns,target):
    for col in columns:
        if col!= target:    
            plt.figure(figsize=(10,6))
            sns.countplot(x=col,hue=target,palette='Set2',data=data)
            plt.title(f'count plot between {col} vs {target}')
            plt.xlabel(col)
            plt.ylabel("count")
            plt.show()
```

##### MODEL SELECTION

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier

from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,roc_auc_score
```

```python
x= df.drop(columns=['booking_status'],axis = 1)
y = df['booking_status']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
models = {
    'random forest' : RandomForestClassifier(random_state=42),
    'gradient boost' : GradientBoostingClassifier(random_state=42),
    'ada boost' : AdaBoostClassifier(random_state=42),
    'xgb' : XGBClassifier(),
    'lightgbm' : LGBMClassifier(random_state=42),
    'd tree' : DecisionTreeClassifier(random_state=42),
    'logistic' : LogisticRegression(random_state=42),
    'svc' : SVC(),
    'gaussian NB' : GaussianNB(),
    'knn' : KNeighborsClassifier()
}
```

```python
results = {
    'model' : [],
    'accuracy_score' : [],
    'precision_score': [],
    'recall_score' : [],
    'f1_score' : [],
    'roc_auc' : []
}
```

```python
for model, classif in models.items():
    classif.fit(x_train,y_train)
    y_pred = classif.predict(x_test)

    accuracy = accuracy_score(y_test,y_pred)
    precision = precision_score(y_test,y_pred)
    recall = recall_score(y_test,y_pred)
    f1 = f1_score(y_test,y_pred)
    roc = roc_auc_score(y_test,y_pred)


    results['model'].append(model)
    results['accuracy_score'].append(accuracy)
    results['precision_score'].append(precision)
    results['recall_score'].append(recall)
    results['f1_score'].append(f1)
    results['roc_auc'].append(roc) 
```

```python
results_df = pd.DataFrame(results)
```

```python
results_df.sort_values(by=['recall_score','f1_score','roc_auc','precision_score','accuracy_score'],ascending=False)
```

Overall Random Forest appears to be the most balanced among all the models

##### HYPERPARAMETER TUNING

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
from sklearn.ensemble import RandomForestClassifier 
```

```python
x= df.drop(columns=['booking_status'],axis = 1)
y = df['booking_status']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
params = {
    'n_estimators' : randint(40,400),
    'max_depth' : randint(5,50),
    'min_samples_split' : randint(2,10),
    'min_samples_leaf' : randint(1,5),
    'max_features' : ['sqrt','log2'],
    'bootstrap' : [True,False]
}
```

```python
rf_best = RandomForestClassifier(random_state=42)
```

```python
random_search = RandomizedSearchCV(
    estimator=rf_best,
    param_distributions=params,
    n_iter=10,
    cv=5,
    verbose=2,
    random_state=42,
    scoring='accuracy'
)
```

```python
random_search.fit(x_train,y_train)
```

```python
random_search.best_params_
```

```python
best_model = random_search.best_estimator_
```

```python
y_pred = best_model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,y_pred)
precision = precision_score(y_test,y_pred)
recall = recall_score(y_test,y_pred)
f1 = f1_score(y_test,y_pred)
roc = roc_auc_score(y_test,y_pred)
```

```python
print(f'accuracy:{accuracy}')
print(f'precision:{precision}')
print(f'recall : {recall}')
print(f'f1:{f1}')
print(f'roc:{roc}')
```
